```python
import pandas as pd

# Load the data from the URL
df = pd.read_csv(
    'https://www.amfiindia.com/spages/NAVAll.txt', 
                sep=';'
    )

# Print the first few rows of the DataFramedf.head()
```

```python
df.head()
```

```python
df = df.loc[df.isnull().mean(axis=1) < 0.5]
```

```python
df = df.reset_index().drop("index",axis=1)
```

```python
df['Date'] = pd.to_datetime(df['Date'], format='%d-%b-%Y')
df['Date_Year'] = df['Date'].dt.year
df['Date_Month'] = df['Date'].dt.month
df['Date_Day'] = df['Date'].dt.day
```

```python
df = df.loc[~((df["Date_Day"] != 20) & (df["Date_Month"] != 12) & (df["Date_Year"] != 2022))]
```

```python
URL = "https://www.nseindia.com/api/equity-stockIndices?index=NIFTY%2050"
```

```python
import requests
import pandas as pd
```

```python
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/108.0.0.0 Safari/537.36'}
```

```python
resp = requests.get(URL,headers=headers)
```

```python
resp
```

```python
stocks = resp.json()['data'][1:]
```

```python
df = pd.DataFrame.from_records(stocks)
```

```python
df.to_csv("price.csv",index=False)
```
